```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv("gas_prices.csv")
gas.head()
```

```python
plt.title('Gas Prices over time($)')
plt.plot(gas.Year, gas.USA, '.r-', label = "USA")
plt.plot(gas.Year, gas.Canada, '.g-', label = "Canada")
plt.plot(gas.Year, gas['South Korea'],'.b-', label = "South Korea")
plt.xticks(gas.Year[::3])
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,10))

for i in gas:
    if i!='Year':
        plt.plot(gas['Year'], gas[i], '.-', label=''+i)
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Price($)')
plt.legend()
plt.show()
```
