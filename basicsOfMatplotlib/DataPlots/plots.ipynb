```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 3D Plots

```python
ax = plt.axes(projection="3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X)*np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='Spectral')
ax.set_title("3D Plots")
plt.savefig('demo.png')
plt.show()
```

```python
ax = plt.axes(projection="3d")
x = np.random.random(50)
y = np.random.random(50)
z = np.random.random(50)

ax.scatter(x, y, z)

plt.show()
```

### Subplots

```python
x = np.arange(100)
fig, axs = plt.subplots(2, 2)

axs[0][0].plot(x, np.sin(x))
axs[0][0].set_title("Sin Wave")

axs[0][1].plot(x, np.cos(x))
axs[0][1].set_title("Cosine Wave")

axs[1][0].plot(x, np.random.random(100))
axs[1][0].set_title("Random Function")

axs[1][1].plot(x, np.log(x))
axs[1][1].set_title("Log Function")
fig.suptitle("Different Types of Plot")
plt.tight_layout()
plt.savefig("Fig.png", dpi=300, transparent=True)

plt.show()
```

### Scatter Plot

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
plt.scatter(x_data, y_data, c="r", marker="x")
plt.show()
```

### Line Plot

```python
years = [2000 + x for x in range(16)]
weights = [25, 50, 90, 42, 51, 32, 95, 41, 32, 10, 60, 54, 87, 84, 22, 68]
height = [12, 15, 20, 21, 15, 52, 15, 25, 21, 26, 23, 41, 22, 12, 45, 12]
```

```python
plt.plot(weights, label="Stock A")
plt.plot(height, label="Stock C")

plt.legend(loc="upper left")
```

```python
plt.plot(years, weights, "--r", lw=3)
plt.show()
```

### Bar Chart

```python
lang = ["C++", "C#", "Python", "Java", "Go"]
votes = [20, 40, 105, 20, 50]
plt.bar(lang, votes, color="#abcdef", align="center", width=0.5, edgecolor="red", lw=2)
plt.show()
```

### Histogram

```python
plt.hist(x_data, cumulative=True, color="gray")
plt.show()
```

### PieChart

```python
explode = [0.1, 0, 0, 0, 0]
plt.pie(votes, labels=lang, explode=explode, autopct="%.2f%%", startangle=0)
plt.show()
```
