### Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

### Unique Sentiment, Emotion, Data Source, Rating

```python
print(df['Sentiment'].unique())
print(df['Emotion'].unique())
print(df['Data Source'].unique())
```

```python
plt.hist(df['Emotion'])
plt.show()
```

```python
labels = df.Emotion.unique()
sizes = []

for label in labels:
    sizes.append(len(df[df.Emotion == label]))

plt.pie(sizes, labels=labels, autopct='%.2f%%')
plt.show()
```
